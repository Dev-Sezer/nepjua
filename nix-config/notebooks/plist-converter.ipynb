```python

import xml.etree.ElementTree as ET

def parse_xml(xml_string):
    root = ET.fromstring(xml_string)
    return root

tree = parse_xml(plist_config)
```

```python
dictionary = {}
for child in tree.iter('key'):
    key = child.text
    value = next(child.itertext())
    if value == 'true':
        value = True
    elif value == 'false':
        value = False
    elif value.startswith('<data>'):
        value = value[7:-8]
        value = bytes.fromhex(value).decode('utf-8', 'ignore')
    elif value.startswith('<string>'):
        value = value[8:-9]
    elif value.startswith('<array>'):
        value = []
        for array_child in child.iter('string'):
            value.append(array_child.text)
    elif value.startswith('<dict>'):
        value = {}
        for dict_child in child.iter('key'):
            dict_key = dict_child.text
            dict_value = next(dict_child.itertext())
            if dict_value == 'true':
                dict_value = True
            elif dict_value == 'false':
                dict_value = False
            elif dict_value.startswith('<data>'):
                dict_value = dict_value[7:-8]
                dict_value = bytes.fromhex(dict_value).decode('utf-8', 'ignore')
            elif dict_value.startswith('<string>'):
                dict_value = dict_value[8:-9]
            elif dict_value.startswith('<integer>'):
                dict_value = int(dict_value[10:-11])
            value[dict_key] = dict_value
    elif value.startswith('<integer>'):
        value = int(value[10:-11])
    dictionary[key] = value
    
dictionary
```
